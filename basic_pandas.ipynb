```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from datetime import datetime
```

```python
%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = (10,5)
```

```python
saved_style_state = matplotlib.rcParams.copy()
```

```python
filepath = "squirrel_data.xlsx"

squirrelnyc_data = pd.read_excel(filepath)
print("done")
```

```python
squirrelnyc_data.sample(3023)
```

```python
fig, ax = plt.subplots()
ax.set_title('Squirrel Primary Color')
ax.set_xlabel('Fur Color')
ax.set_ylabel('Frequency of Type of Fur Color')
squirrelnyc_data["Primary Fur Color"].value_counts().plot(kind="bar", facecolor="purple")
```

```python
fig, ax = plt.subplots()
ax.set_title('Location of Squirrels')
squirrelnyc_data["Location"].value_counts().plot(kind="pie")
```

```python
fig, ax = plt.subplots()
ax.set_title('Squirrel Indifference to Human Presence')
ax.set_xlabel('')
ax.set_ylabel('Frequency of Indifferences')
squirrelnyc_data["Indifferent"].value_counts().plot(kind="bar", facecolor="lightblue")
```

```python
fig, ax = plt.subplots()
ax.set_title('Number of Squirrel Sighted in a Discrete Sighting Session')
ax.set_xlabel('Number of Squirrels Spotted')
ax.set_ylabel('Number of Squirrel Spotted')
squirrelnyc_data["Hectare Squirrel Number"].value_counts().plot(kind="bar", facecolor="magenta")
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Running', 'Chasing', 'Climbing', 'Eating', 'Foraging']
values_group1 = [730, 279, 658, 760, 1435]  
values_group2 = [2293, 2744, 2365, 2263, 1588]  

x = np.arange(len(categories))

bar_width = 0.4

fig, ax = plt.subplots()

ax.bar(x - bar_width/2, values_group1, bar_width, label='True')
ax.bar(x + bar_width/2, values_group2, bar_width, label='False')

ax.set_ylabel('Frequency of Squirrel')
ax.set_title('Squirrel Activities')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Other','Runs from dog', 'Runs from dog', 'Curious', 'Stared', 'Observant', 'Avoided', 'Ran up Tree', 'Fenced off']
values = [198, 11, 9, 6, 4, 2, 3, 5, 2]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Other Interactions')
plt.xticks(rotation=45) 
plt.tight_layout()  
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_title('Shift Times of Squirrel Spottings')
ax.set_xlabel('Number of Squirrels Spotted')
ax.set_ylabel('Shift')
squirrelnyc_data["Shift"].value_counts().plot(kind="barh")
```

```python
fig, ax = plt.subplots()
ax.set_title('Date of Number of Squirrels seen')
ax.set_xlabel('')
ax.set_ylabel('Frequency of Fur Color')
squirrelnyc_data["Date"].value_counts().plot(kind="pie")
```

```python
excel_file_path = 'squirrel_data.xlsx'
data = pd.read_excel(excel_file_path)

latitudes = data['X']
longitudes = data['Y']

plt.figure(figsize=(10, 6))
plt.scatter(longitudes, latitudes, color='blue', marker='o')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Longitude and Latitude')

plt.grid(True)
plt.tight_layout()
plt.show()
```
